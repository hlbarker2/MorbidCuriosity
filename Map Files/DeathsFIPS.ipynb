```python
import pandas as pd
```

```python
geojson = pd.read_csv("us_county_geojson.csv")
```

```python
geojson.head(50)
```

```python
geojson["Full County"] = geojson["county"] + ", " + geojson["state_code"]
```

```python
del geojson["county"]
del geojson["state_code"]
geojson.head(50)
```

```python
fips = pd.read_csv("Fips_Counties.csv")
fips["FIPS"] = fips["FIPS"].apply(lambda x: '{0:0>5}'.format(x))
fips.head(50)
```

```python
fips["Full County"] = fips["Name"] + ", " + fips["State"]
del fips["Name"]
del fips["State"]
fips.head(50)
```

```python
deaths = pd.read_csv("Deaths.csv")
deaths["County Code"] = deaths["County Code"].apply(lambda x: '{0:0>5}'.format(x))
deaths.head(25)
```

```python
del deaths["County"]
deaths.head(25)
```

```python
merge1 = pd.merge(geojson, fips, on="Full County")
merge1.head(25)
```

```python
deaths.rename(columns={'County Code':'FIPS'}, inplace=True)
deaths.head(25)
```

```python
merge2 = pd.merge(merge1, deaths, on="FIPS")
merge2.head(25)
```

```python
merge2.to_csv("DeathsFIPS.csv", index=False, header=True)
```
