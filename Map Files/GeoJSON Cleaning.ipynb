```python
import json
import geojson
from pprint import pprint
import pandas as pd
```

```python
usgeojson = {
    "type": "FeatureCollection",
    "features": []
}
```

```python
death = pd.read_csv("AllDeath.csv")
death['FIPS'] = death['FIPS'].apply(lambda x: '{0:0>5}'.format(x))
death["Population"] = death["Population"].fillna("0")
death["Total per enrollee (age, sex, race, price adjusted)"] = death["Total per enrollee (age, sex, race, price adjusted)"].fillna("0")
death["Total per enrollee (age, sex, race, price adjusted)"] = death["Total per enrollee (age, sex, race, price adjusted)"].str.replace(",", "")
death["Population"] = death["Population"].astype('int64')
death["Total per enrollee (age, sex, race, price adjusted)"] = death["Total per enrollee (age, sex, race, price adjusted)"].astype('float')
death = death.fillna("Not Available")
death.head(15)
```

```python
import numpy as np
death["Population"] = np.where(death.Population > 100000, 100000, death.Population)
death["Total per enrollee (age, sex, race, price adjusted)"] = np.where(death["Total per enrollee (age, sex, race, price adjusted)"] > 15000, 15000, death["Total per enrollee (age, sex, race, price adjusted)"])
```

```python
death_clean = death[["FIPS", "LOCATION", "Population", "Total_Death_Rate", "Cancer_Death_Rate", "Heart_Death_Rate",
               "Stroke_Death_Rate", "LowerResp_Death_Rate", "Accident_Death_Rate", "Life Expectancy", 
              "Total per enrollee (age, sex, race, price adjusted)"]]
death_clean.head(100)
```

```python
states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware",
"District_of_Columbia", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas",
"Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri",
"Montana", "Nebraska", "Nevada", "New_Hampshire", "New_Jersey", "New_Mexico", "New_York", "North_Carolina",
"North_Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode_Island", "South_Carolina", "South_Dakota",
"Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West_Virginia", "Wisconsin", "Wyoming"]
```

```python
for state in states:
    with open(f'GeoJSON/{state}.geo.json') as f:
        features = json.load(f)["features"]
        for foo in features:
            try:
                fips = foo['properties']["FIPS"]
                deathdata = death_clean.loc[death_clean["FIPS"] == fips, :]
                deathdict = deathdata.to_dict(orient = 'records')[0]
                foo["properties"] = deathdict
                usgeojson["features"].append(foo)
            except:
                print(f'An error occured for FIPS #: {fips}')
```

```python
with open('us.geo.json', 'w') as fp:
    geojson.dump(usgeojson, fp)
```
