```python
#Import dependencies
import pandas as pd
```

```python
#Import zips CSV
zips_df = pd.read_csv('Datasets/ziptofips.csv')
zips_df.head()
```

```python
#Remove Unnamed column
zips_df = zips_df[['ZipCodes', 'FIPS']]
zips_df.head()
```

```python
#Add leading zeros to FIPS
zips_df['FIPS'] = zips_df['FIPS'].apply(lambda x: '{0:0>5}'.format(x))
zips_df
```

```python
#Export DF to CSV
zips_df.to_csv('Output/Cleanedziptofips.csv')
```
